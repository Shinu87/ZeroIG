```python
!python test.py --data_path_test_low /content/ZeroIG/data/low --model_test /content/ZeroIG/EXP/Train-20250429-081158/model_epochs/weights_200.pt
```

```python
import os
import numpy as np
from PIL import Image
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Set paths
enhanced_dir = './results/result/'
gt_dir = '/content/ZeroIG/data/high'

# Collect results
psnr_scores = []
ssim_scores = []

# Process each enhanced image
for filename in sorted(os.listdir(enhanced_dir)):
    if not filename.endswith('.png'):
        continue

    # Extract the base name (e.g., from '1_denoise.png' -> '1')
    base_name = filename.split('_')[0]

    enhanced_path = os.path.join(enhanced_dir, filename)

    # Try different GT extensions
    gt_path = None
    for ext in ['.png', '.jpg', '.jpeg']:
        temp_path = os.path.join(gt_dir, base_name + ext)
        if os.path.exists(temp_path):
            gt_path = temp_path
            break

    if gt_path is None:
        print(f"[Warning] GT image not found for: {filename}")
        continue

    # Load images
    enhanced_img = np.array(Image.open(enhanced_path).convert('RGB'))
    gt_img_raw = Image.open(gt_path).convert('RGB')
    gt_img_resized = gt_img_raw.resize((enhanced_img.shape[1], enhanced_img.shape[0]), Image.BICUBIC)
    gt_img = np.array(gt_img_resized)

    # Ensure sizes match
    if enhanced_img.shape != gt_img.shape:
        print(f"[Warning] Size mismatch for {filename}, skipping...")
        continue

    # Calculate PSNR and SSIM
    psnr_val = psnr(gt_img, enhanced_img, data_range=255)
    ssim_val = ssim(gt_img, enhanced_img, data_range=255, channel_axis=-1)

    psnr_scores.append(psnr_val)
    ssim_scores.append(ssim_val)

    print(f"{filename} - PSNR: {psnr_val:.2f}, SSIM: {ssim_val:.4f}")

# Final evaluation
print("\n--- Final Evaluation ---")
if psnr_scores and ssim_scores:
    avg_psnr_lol = np.mean(psnr_scores)
    avg_ssim_lol = np.mean(ssim_scores)
    print(f"Average PSNR: {avg_psnr_lol:.2f} dB")
    print(f"Average SSIM: {avg_ssim_lol:.4f}")
else:
    avg_psnr = None
    avg_ssim = None
    print("No valid comparisons.")
```

```python
import os
import numpy as np
from PIL import Image
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Set paths
enhanced_dir = './results/result/'  # Directory with output images (e.g., denoise/enhance)
gt_dir = '/content/LSRW/Nikon/high'  # Directory with ground truth images

# Collect results
psnr_scores = []
ssim_scores = []

# Process each enhanced image
for filename in sorted(os.listdir(enhanced_dir)):
    if not filename.endswith('.png'):
        continue

    # Extract the base name (e.g., from '3001_denoise.png' -> '3001')
    base_name = filename.split('_')[0]

    enhanced_path = os.path.join(enhanced_dir, filename)

    # Try matching GT with different extensions
    gt_path = None
    for ext in ['.png', '.jpg', '.jpeg']:
        temp_path = os.path.join(gt_dir, base_name + ext)
        if os.path.exists(temp_path):
            gt_path = temp_path
            break

    if gt_path is None:
        print(f"[Warning] GT image not found for: {filename}")
        continue

    # Load and convert images to RGB
    enhanced_img = np.array(Image.open(enhanced_path).convert('RGB'))
    gt_img_raw = Image.open(gt_path).convert('RGB')

    # Resize GT to match enhanced size
    gt_img_resized = gt_img_raw.resize((enhanced_img.shape[1], enhanced_img.shape[0]), Image.BICUBIC)
    gt_img = np.array(gt_img_resized)

    # Ensure sizes match
    if enhanced_img.shape != gt_img.shape:
        print(f"[Warning] Size mismatch for {filename}, skipping...")
        continue

    # Calculate metrics
    psnr_val = psnr(gt_img, enhanced_img, data_range=255)
    ssim_val = ssim(gt_img, enhanced_img, data_range=255, channel_axis=-1)

    psnr_scores.append(psnr_val)
    ssim_scores.append(ssim_val)

    print(f"{filename} - PSNR: {psnr_val:.2f}, SSIM: {ssim_val:.4f}")

# Final average metrics
print("\n--- Final Evaluation ---")
if psnr_scores and ssim_scores:
    avg_psnr = np.mean(psnr_scores)
    avg_ssim = np.mean(ssim_scores)
    print(f"Average PSNR: {avg_psnr:.2f} dB")
    print(f"Average SSIM: {avg_ssim:.4f}")
else:
    print("No valid comparisons.")
```
